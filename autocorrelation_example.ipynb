```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics import tsaplots


x = np.array(range(100))

#Try different sets of data (decending order of autocorrelation)

y = np.array(range(100))
#y = np.array(50*[2,5])
#y = np.array(10*[2,5,2,8,5,31,8,2,46,1])
#y = np.random.random(100)
#y = ...

fig1 = plt.scatter(x,y,c = "black", marker = ".")
plt.xlabel(r"$t$", fontsize = 20)
plt.ylabel("value", fontsize = 20)
plt.show


acf = sm.tsa.stattools.acf(y, nlags = 10, fft=False)
print(acf)

fig2 = tsaplots.plot_acf(y, lags=10, fft=False)
plt.show()
```
